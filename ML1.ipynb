### Question1- Imagine you have a dataset where you have different Instagram features like u sername , Caption , Hashtag , Followers , Time_Since_posted , and likes , now your task is to predict the number of likes and Time Since posted and the rest of the features are your input features. Now you have to build a model which can predict the number of likes and Time Since posted.


```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.multioutput import MultiOutputRegressor
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
data=pd.read_csv("instagram_reach.csv")
```

```python
data
```

```python
data=data.drop("Unnamed: 0",axis=1)
```

```python
data
```

```python
data.isnull().sum()
```

```python
data.dropna(inplace=True)
```

```python
data
```

```python
data['Time since posted']=data['Time since posted'].apply(lambda x: x.replace("hours", ""))
```

```python
data
```

```python
data = pd.get_dummies(data, columns=["USERNAME", "Caption", "Hashtags"])
```

```python
data['Time since posted']=data['Time since posted'].apply(lambda x: x.replace("hours", ""))
```

```python
features = data.drop(columns=["Likes", "Time since posted","S.No"])
target_likes = data["Likes"]
target_time_since_posted = data["Time since posted"]
```

```python
X_train, X_test, y_train_likes, y_test_likes, y_train_time, y_test_time = train_test_split(
    features, target_likes, target_time_since_posted, test_size=0.2, random_state=21
)
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python

model = MultiOutputRegressor(LinearRegression())
```

```python
model.fit(X_train_scaled, pd.concat([y_train_likes, y_train_time], axis=1))

predictions = model.predict(X_test_scaled)
likes_predictions = predictions[:, 0]
time_predictions = predictions[:, 1]

likes_mse = mean_squared_error(y_test_likes, likes_predictions)
likes_mae = mean_absolute_error(y_test_likes, likes_predictions)

time_mse = mean_squared_error(y_test_time, time_predictions)
time_mae = mean_absolute_error(y_test_time, time_predictions)

print("Likes - MSE: {:.2f}, MAE: {:.2f}".format(likes_mse, likes_mae))
print("Time Since Posted - MSE: {:.2f}, MAE: {:.2f}".format(time_mse, time_mae))
```
