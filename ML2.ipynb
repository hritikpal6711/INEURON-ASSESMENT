### Q2-Imagine you have a dataset where you have different features like Age ,Gender , Height , Weight , BMI , and Blood Pressure and you have to classify the people into different classes like Normal , Overweight , Obesity , Underweight , and Extreme Obesity by using any 4 different classification algorithms. Now you have to build a model which can classify people into different classes

```python
import pandas as pd
```

```python
data = pd.read_csv('ObesityDataSet_raw_and_data_sinthetic.csv') #reading dataset
```

```python
data
```

```python
data.isnull().sum() #no null value
```

```python
data['Gender'] = data['Gender'].map({'Male': 0, 'Female': 1})  # Convert categorical variable to numeric
```

```python
def obese(bmi):
    
    if bmi <= 18.5:
        return "Underweight"
    elif bmi <= 24.9:
        return "Healthy weight"
    elif bmi <= 29.9:
        return "Overweight"
    elif bmi <= 40:
        return "Obesity"
    else:
        return "Extreme Obesity"
```

```python
data['obesityLevel']=data["Weight"]/(data['Height']*data['Height'])
```

```python
data['obese or not']=data['obesityLevel'].apply(lambda x: obese(x))
```

```python
data
```

```python
features = data[['Age', 'Gender', 'Height', 'Weight']]   #feature selection
target = data['obese or not']
```

```python
from sklearn.model_selection import train_test_split      #performing train test split

X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression        #importing main 4 classification algorithms
from sklearn.tree import DecisionTreeClassifier   
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
```

```python
lr_model = LogisticRegression()
lr_model.fit(X_train, y_train)

dt_model = DecisionTreeClassifier()
dt_model.fit(X_train, y_train)

rf_model = RandomForestClassifier()
rf_model.fit(X_train, y_train)

svm_model = SVC()
svm_model.fit(X_train, y_train)

from sklearn.metrics import classification_report

# Make predictions on the test set
lr_predictions = lr_model.predict(X_test)
dt_predictions = dt_model.predict(X_test)
rf_predictions = rf_model.predict(X_test)
svm_predictions = svm_model.predict(X_test)

# Evaluate the models
lr_report = classification_report(y_test, lr_predictions)
dt_report = classification_report(y_test, dt_predictions)
rf_report = classification_report(y_test, rf_predictions)
svm_report = classification_report(y_test, svm_predictions)

print("Logistic Regression:\n", lr_report)
print("Decision Tree:\n", dt_report)
print("Random Forest:\n", rf_report)
print("Support Vector Machine:\n", svm_report)
```

After interpreting the result, it is clearly seen that Decision Tree classifier gives better result and also gives better performance in respect of the given dataset.
